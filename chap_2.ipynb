Indexing data

```python
A = np.array(np.arange(16)).reshape((4,4))
A
```

```python
A[1,2]
```

```python
#select multiple rows at a time
A[[1,2]]
```

```python
#select first and third columns
A[:, [0,2]]
```

```python
#let's say we want to select a submatrix made up of the second and fourth rows as well as the first and third columns. 
#first, create submatrix by subsetting rows of A
#second, make a further submatrix by subsetting its columns.

A[[1,3]][:,[0,2]]
```

```python
#more convenient way is the np.ix_() to create an intermediate mesh object.
idx = np.ix_([1,3], [0,2,3])
A[idx]
```

```python
#OR we can use slices. 
# [1:4:2] captures second and fourth items of sequence with step 2..
# [0:3:2] captures first and third items with step of 2. 

A[1:4:2,0:3:2]
```

Boolean indexing        

```python

keep_rows = np.zeros(A.shape[0], bool)
keep_rows
```

```python
keep_rows[[1,3]] = True
keep_rows
```

```python
np.all(keep_rows == np.array([0,1,0,1]))
#checks to see if all the entries of an array are True. np.any() can check if any entries in an array are true. 
```

```python
#but, these two arrays index different sets of rows! booleans and integers are treated differently in numpy.
A[np.array([0,1,0,1])]
```

```python
A[keep_rows]
```

```python
#now we use the np.ix() function to create a mesh containing the second and fourth rows, and the first, third, and fourth columns.
#this time we apply to booleans rather than lists. 

keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0,2,3]] = True
idx_bool = np.ix_(keep_rows, keep_cols)
A[idx_bool]
```

```python
#you can also mix a list with an array of Booleans in the argument to idx_()
idx_mixed = np.ix_([1,3], keep_cols)
A[idx_mixed]
```

Selecting rows and columns

```python
#use Auto.columns to check variable names
Auto.columns
```

```python
#remember, as with numpy, the first argument to the [] method is always applied to the rows of the array.
#Similalry, passing in a slice to [] method creates a df whose ROWS are determined by the slice

Auto[:3]
```

```python
#you can use booleans to subset. 

idx_80 = Auto['year']   > 80
Auto[idx_80].shape # we see there are only 59 rows with car age >80
```

```python
#if we past list of strings to [] we get df with the set fo columns from that list. 

Auto[['mpg', 'horsepower']]
```

```python
#notice that the index is labeled with ints. We can reset the index to be the column of name. 

Auto_re = Auto.set_index('name')
Auto_re
```

```python
Auto_re.columns
```

```python
#now we can select rows by name instead of the loc..

rows = ['amc rebel sst', 'ford torino']
Auto_re.loc[rows]
```

```python
#as alternative to using the index name, we could retrieve the 4th and 5th rows using iloc[]

Auto_re.iloc[[3,4]]
```

```python
#or grab the columns!
Auto_re.iloc[:, [0,2,3]]
```

```python
#extract the 4th and 5th rows as well as 1st, 3rd, 4th columns with single call to iloc.
Auto_re.iloc[[3,4], [0,2,3]]
```

```python
#the index entry need not be unique. there are multiple entries for call ford galaxie 500
Auto_re.loc['ford galaxie 500', ['mpg', 'origin']]
```

```python
#more on selecting rows/columns

idx_80 = Auto_re['year'] > 80
Auto_re.loc[idx_80, ['weight', 'origin']]
```

```python
#or we can use a lambda function
#lambda creates a function that takes signle argument (here df) and returns df['year'] > 80
Auto_re.loc[lambda df: df['year']>80, ['weight', 'origin']]
```

```python
Auto_re.loc[lambda df: (df['year'] > 80) & (df['mpg'] > 30), 
            ['weight', 'origin']]

#& computes element-wise and operation.
```

```python
Auto_re.loc[lambda df: (df['displacement'] <300) & (df.index.str.contains('ford')
                                                    | df.index.str.contains('datsun')),
                                                    ['weight', 'origin']
                                                    
                                                    ]

#here the pipe is an OR operator.
```

```python
#using loops

total = 0
for value in [3,2,19]:
    total += value
print('total is: {0}'.format(total))
```

```python
total = 0
for value in [3,2,19]:
    for weight in [3,6,10]:
        total += value*weight

print('total is: {0}'.format(total))
```

```python
#perhaps we want to sum over (value, weight) pairs. 

total = 0
for value, weight in zip([2,3,19], 
                         [.2,.3,.5]):
    total += weight * value
print('weighted average is: {0}'.format(total))
```

String formatting

```python
rng = np.random.default_rng(1)
A = rng.standard_normal((127, 5))
M = rng.choice([0, np.nan], p= [.8, .2], size = A.shape)
A += M
D = pd.DataFrame(A, columns=['food', 'bar', 'pickle', 'snack', 'popcorn'])
D[:3]
```

```python
for col in D.columns:
    template = 'Column "{0}" has {1:.2%} missing values'
    print(template.format(col, 
                          np.isnan(D[col]).mean()))
```
